pyplot provides a user-friendly interface for plotting

```python
import matplotlib.pyplot as plt

x = [2023,2024,2025,2026]
y = [15,25,30,20]

# first arg for points on x-axis
plt.plot(y)

# default behaviour is to start at 0

plt.show()

```

```python
plt.plot(x,y)
plt.show()
```

Numpy they are faster than list and provide more functionality

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5])
y = np.arange(6,11)
#matplot needs same length for x and y
plt.plot(x,y)
plt.show()
```
