Histogram : A visual representation of the distribution of quantitative data.
they group values into bins (intervals)
and counts how many fall in each range

```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
scores = np.random.normal(loc=80, scale = 10, size=100)
#normal distribution
#loc is the location of the center i.e. the median of our data
#scale : how far scores deviate from center . the higher the number the greater the deviation
#size : how many numbers we want to generate

plt.hist(scores)
plt.show()
```

```python
#to ensure no score goes below 0 or above 100 we can use clip function

scores = np.clip(scores,0,100)
plt.hist(scores)
plt.show()
```

more customization:
we can set the no. of bins we want as another keyword argument within hist function

```python
plt.hist(scores, bins=12,
                 color= "lightgreen",
                 edgecolor = "black")
plt.title("Exam Scores")
plt.xlabel("Score")
plt.ylabel("No. of students")
plt.show()
```
