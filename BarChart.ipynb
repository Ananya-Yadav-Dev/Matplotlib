Barchart = compare categories of data by representing each category with a bar

```python
import matplotlib.pyplot as plt
import  numpy as np
```

```python
categories = ["Grains","Fruits","Vegetables","Protein","Diary","Sweets"]
values = np.array([4,3,2,5,3,1])

plt.bar(categories,values)
plt.show()
```

```python
label_design = dict(fontsize=14, color='blue', fontweight='bold')


plt.bar(categories,values)
plt.title("Daily consumption",fontsize=25,fontweight='bold')

plt.xlabel("Food Categories",**label_design)
plt.ylabel("Quantity",**label_design)
```

```python
plt.bar(categories,values,color=["skyblue","orange"])
```

```python
plt.barh(categories,values,color=["skyblue","orange"])
```
