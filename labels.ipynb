```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([2023,2024,2025,2026])
y1 = np.array([15,25,30,20])
y2 = np.array([17,23,38,5])

plt.title("Class Size")

plt.plot(x1,y1)
plt.plot(x1,y2)

plt.show()
```

```python
plt.plot(x1,y1)
plt.plot(x1,y2)
plt.title("Class size",fontsize=20,family="Arial",fontweight="bold",color="green")
plt.show()
```

```python
#label for x-axis

plt.plot(x1,y1)
plt.plot(x1,y2)

plt.title("Class size",
          fontsize=25,
          family="Arial",
          fontweight="bold",
          color="#2d4cfc")

plt.xlabel("Year",
           fontsize=20,
           family="Arial",
           fontweight="bold",
           color="blue")

plt.ylabel("Year",
           fontsize=20,
           family="Arial",
           fontweight="bold",
           color="blue")

plt.show()
```

```python
# Ticks only for x-axis
plt.plot(x1,y1)
plt.plot(x1,y2)

plt.title("Class size",
          fontsize=25,
          family="Arial",
          fontweight="bold",
          color="#2d4cfc")

plt.xlabel("Year",
           fontsize=20,
           family="Arial",
           fontweight="bold",
           color="blue")

plt.ylabel("Year",
           fontsize=20,
           family="Arial",
           fontweight="bold",
           color="blue")

# Set x-axis ticks to only show your data points
plt.xticks(x1)

plt.tick_params(axis="both",
                color="red",
                labelsize=15)


plt.show()
```
