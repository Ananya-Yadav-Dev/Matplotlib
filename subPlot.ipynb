Figure = a single canvas
Ax = A single plot (subplot)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.subplots(2,2)
```

```python
figure , axes = plt.subplots(2,2)
# axes is technically a numpy array

x = np.array([1,2,3,4,5])

axes[0,0].plot(x,x*2,color="red")
axes[0,0].set_title("x*2")

plt.show()
```

```python
figure , axes = plt.subplots(2,2)

axes[0,0].plot(x,x*2,color="red")
axes[0,0].set_title("x*2")

axes[0,1].bar(x,x**2,color="blue")
axes[0,1].set_title("x**2")

axes[1,1].plot(x,x**3,color="green")
axes[1,1].set_title("x**3")

axes[1,0].bar(x,x**4,color="yellow")
axes[1,0].set_title("x**4")

#plots overlapping 
plt.tight_layout()

plt.show()
```
