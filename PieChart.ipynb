Pie Chart : circular chart divided into slices to show percentages of the total. Good for visualizing distribution among categories

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
categories = ["Freshmen","Sophomores","Juniors","Seniors"]
values = np.array([50, 75, 60, 90])

plt.pie(values)

plt.show()
```

```python
plt.pie(values, labels=categories,
                autopct='%1.1f%%') #keyword arg auto percentage .Also.1 to limit the digits after decimal
#we use double % sign as one is used as format specifier
plt.show()
```

```python
# set colors
colors = ['gold', 'red', 'skyblue', 'green']
plt.pie(values, labels=categories,
                autopct='%1.1f%%',
                colors=colors) 
plt.title("Class Distribution")
plt.show()
```

```python
# to highlight a pie slice, we can use the 'explode' parameter.
plt.pie(values, labels=categories,
                autopct='%1.1f%%',
                colors=colors,
                explode=(0.1, 0, 0, 0))
plt.show()
```

```python
plt.pie(values, labels=categories,
                autopct='%1.1f%%',
                colors=colors,
                explode=(0.1, 0, 0, 0),
                shadow=True,
                startangle=90)
plt.show()
```
