Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,5)
y = np.arange(6,10)

plt.plot(x, y, marker = "v")

plt.show()
```

```python
plt.plot(x, y, marker = "o",ms = 10)
plt.show()
```

```python
plt.plot(x, y, marker = ".", 
               markersize=30,
               markerfacecolor="cyan")
plt.show()
```

```python
plt.plot(x, y, marker = ".", 
               ms=30,
               mfc="cyan",
               markeredgecolor="cyan")
plt.show()
```

```python
plt.plot(x, y, marker = ".", 
               ms=30,
               mfc="cyan",
               markeredgecolor="cyan",
               linestyle = "--")
plt.show()
```

```python
plt.plot(x, y, marker = ".", 
               ms=30,
               mfc="cyan",
               mec="cyan",
               linestyle = "dotted")
plt.show()
```

```python
plt.plot(x, y, marker = ".", 
               ms=30,
               mfc="cyan",
               markeredgecolor="cyan",
               linestyle="None")
plt.show()
```
