Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,5)
y = np.arange(6,10)

plt.plot(x, y, marker = "v")

plt.show()
```

```python
plt.plot(x, y, marker = "o",ms = 10)
plt.show()
```

```python
plt.plot(x, y, marker = ".", 
               markersize=30,
               markerfacecolor="cyan")
plt.show()
```

```python
plt.plot(x, y, marker = ".", 
               ms=30,
               mfc="cyan",
               markeredgecolor="cyan")
plt.show()
```

```python
plt.plot(x, y, marker = ".", 
               ms=30,
               mfc="cyan",
               markeredgecolor="cyan",
               linestyle = "--")
plt.show()
```

```python
plt.plot(x, y, marker = ".", 
               ms=30,
               mfc="cyan",
               mec="cyan",
               linestyle = "dotted")
plt.show()
```

```python
plt.plot(x, y, marker = ".", 
               ms=30,
               mfc="cyan",
               markeredgecolor="cyan",
               linestyle="None")
plt.show()
```

```python
plt.plot(x, y, marker = ".", 
               ms=30,
               mfc="cyan",
               mec="cyan",
               ls="solid",
               linewidth=3,
               color="magenta")
plt.show()
```

```python
x1 = np.array([2023,2024,2025,2026])
y1 = np.array([50,100,150,200])
y2 = np.array([30,60,90,120])

plt.plot(x1,y1, marker = "o", ms=10, mfc="red", mec="red", ls="--", linewidth=2, color="blue")
plt.plot(x1,y2, marker = "s", ms=10, mfc="green", mec="green", ls=":", linewidth=2, color="orange")
plt.show()
```

```python
# rather passing as argument. create a dictionary of the properties and pass it using **kwargs

line_style = dict(marker = "o", ms=10, mfc="green", mec="green", ls=":", linewidth=2, color="orange")
plt.plot(x1,y1, **line_style) #unpacking the line_style dictionary
plt.plot(x1,y2,**line_style)
plt.show()
```

```python
line_style = dict(marker = "o", ms=10, mfc="green", mec="green", ls=":", linewidth=2)
plt.plot(x1,y1, **line_style) #unpacking the line_style dictionary
plt.plot(x1,y2,color="blue",**line_style)
plt.show()
```
