Grid Lines 
grid() = Helps males plots easier to read by adding reference lines.

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [5,10,15,20,25]

plt.grid() #grid lines with no customization

plt.plot(x,y)
plt.show()
```

```python
#if only want single axis grid lines
plt.grid(axis="x",
         linewidth=2,
         color = "lightgray",
         ls="dashdot")
plt.plot(x,y)
```
