```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("world_population.csv")
df
```

```python
continent_count = df["Continent"].value_counts()
continent_count
```

```python
plt.barh(continent_count.index,continent_count.values)
```

```python
# i want max value at top
continent_count = df["Continent"].value_counts(ascending=True)
plt.barh(continent_count.index,continent_count.values,edgecolor = "black")
plt.title("Continent vs Count")
plt.ylabel("Continent")
plt.xlabel("Count")

#to make everything fits call tight layout function
plt.tight_layout()
plt.show()
```
